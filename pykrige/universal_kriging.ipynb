```python
import default_values as dv
import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
from pykrige import UniversalKriging
```

```python
UK = UniversalKriging(
    dv.dataframe_with_values["longitude"],
    dv.dataframe_with_values["latitude"],
    dv.dataframe_with_values["argila"],
    variogram_model="linear",
    drift_terms=["regional_linear"]
)
```

```python
z, ss = UK.execute("grid", dv.xx_grid, dv.yy_grid)
plt.imshow(z)
plt.show()
```
